```python
#Libraries needed
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing Data
Data = pd.read_csv("C:/Users/lenovo/Downloads/HBI1.csv")
```

```python
#Scatter plot between Number of Urban Hospitals and it's Occupancy in Indian states
plt.scatter(Data['NumUrbanHospitals_NHP18'],Data['NumUrbanBeds_NHP18'],c='green')
plt.title("Scatter Plot Between Number of Urban Hospitals and it's Occupancy")
plt.xlabel("Number of Hospitals")
plt.ylabel("No of Beds")
plt.show()
```

```python
#Linear Relationship between Number of Urban Hospitals and their Occupancy
#India making a stable arrangement for healthcare options.
```

```python
#Scatter plot between number of Urban Hospitals vs The number of Rural Hospitals in states of India
plt.scatter(Data['NumUrbanHospitals_NHP18'],Data['NumRuralHospitals_NHP18'],c='red')
plt.title("Scatter Plot Between Number of Urban Hospitals and Number of Rural Hospitals")
plt.xlabel("Number of Urban Hospitals")
plt.ylabel("Number of Rural Hospitals")
plt.show()
```

```python
#It is clear that India is focusing more on Rural Health.
#Due to COVID19,people of old age are at high risk and their percentage is more in rural areasso it is good to have more health care options in rural areas.
```

```python
#Scatter plot between number of District Hospitals vs The number of SubDistrict Hospitals in states of India
plt.scatter(Data['NumDistrictHospitals_HMIS'],Data['NumSubDistrictHospitals_HMIS'],c='green')
plt.title("Scatter Plot Between Number of District Hospitals and Number of SubDistrict Hospitals")
plt.xlabel("Number of District Hospitals")
plt.ylabel("Number of SubDistrict Hospitals")
plt.show()
```

```python
#India maintaining a balance between it's HealthCare options whether it be District place or a SubDistrict area.
```

```python
#Scatter plot between number of Total Public Health Facilities vs The number of Public Beds in states of India
plt.scatter(Data['TotalPublicHealthFacilities_HMIS'],Data['NumPublicBeds_HMIS'],c='red')
plt.title("Scatter Plot Between Total Number of Public Health Facilities and Number of Public Beds")
plt.xlabel("Total Public Health Facilities")
plt.ylabel("Number of Public Beds")
plt.show()
```

```python
#Again quite a linear Relationship.
#Again India maintaining a balance according to the needs of health facilities.
```

```python
#Scatter plot between number of Rural Hospitals vs The number of Rural Beds in states of India
plt.scatter(Data['NumRuralHospitals_NHP18'],Data['NumRuralBeds_NHP18'],c='green')
plt.title("Scatter Plot Between Number of Rural Hospitals and Number of Rural Beds")
plt.xlabel("Number of Urban Hospitals")
plt.ylabel("Number of Rural Beds")
plt.show()
```

```python
#Again a linear relationship
#India maintaining a balance between the occupancy of the rural hospitals in it's various states.
```

```python
#Scatter plot between number of Primary Health Centers vs The number of Community Health Centers in states of India
plt.scatter(Data['NumPrimaryHealthCenters_HMIS'],Data['NumCommunityHealthCenters_HMIS'],c='red')
plt.title("Scatter Plot Between Number of Primary Health Centers and Number of Community Health Centers")
plt.xlabel("Number of Primary Health Centers")
plt.ylabel("Number of Community Health Centers")
plt.show()
```
